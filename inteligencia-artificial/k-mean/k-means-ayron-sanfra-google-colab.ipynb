Este trabalho acadêmico descreve a aplicação do modelo K-Means, uma técnica de aprendizado de máquina não supervisionado, para a análise e agrupamento de dados. O código em Python apresentado neste estudo utiliza o conjunto de dados Iris e demonstra como aplicar o algoritmo K-Means para identificar e visualizar clusters de amostras de plantas com base em suas características.

##Introdução
O modelo K-Means é um dos algoritmos de clustering mais amplamente utilizados em tarefas de agrupamento de dados. Ele é uma técnica de aprendizado de máquina não supervisionado que busca dividir um conjunto de dados em grupos ou clusters com base na similaridade entre as amostras. O algoritmo é particularmente útil para segmentar dados em grupos com características semelhantes, facilitando a análise exploratória e a tomada de decisões.

##Metodologia
###<strong>Conjunto de Dados</strong>
Neste estudo, utilizamos o conjunto de dados Iris, que contém informações sobre três espécies de flores (Setosa, Versicolor e Virginica) com base em quatro características: comprimento da sépala, largura da sépala, comprimento da pétala e largura da pétala.

####<strong>Algoritmo K-Means</strong>
O algoritmo K-Means é aplicado da seguinte forma:

1. Carregar o conjunto de dados Iris usando a biblioteca Pandas.
2. Selecionar as características a serem utilizadas para o agrupamento, que incluem comprimento da sépala, largura da sépala, comprimento da pétala e largura da pétala.
3. Definir o número de clusters (neste caso, definimos como 3, pois existem três espécies de flores no conjunto de dados).
4. Instanciar o modelo K-Means.
5. Treinar o modelo K-Means com as características do conjunto de dados.
6. Obter os rótulos dos clusters para cada amostra.
7. Visualizar os clusters no espaço de características (com as características de comprimento da sépala e largura da sépala).

###Análise e Visualização
Além de aplicar o modelo K-Means, este estudo também realiza as seguintes análises:

 - Cálculo do índice de silhueta para avaliar a qualidade do agrupamento.
 - Cálculo da distância quadrática intra-cluster, que representa a compactação dos clusters.
 - Cálculo da distância quadrática inter-cluster, que representa a separação entre os clusters.
 - Também é realizada a visualização das estatísticas de cada cluster, incluindo o número de amostras em cada cluster e as coordenadas dos centróides.

##Resultados
Os resultados obtidos demonstram a capacidade do modelo K-Means de agrupar efetivamente as amostras de flores Iris em três clusters com base nas características selecionadas. A análise do índice de silhueta indica a qualidade dos clusters formados, enquanto as métricas de distância intra e inter-cluster fornecem informações sobre a compactação e a separação dos clusters.

A visualização dos clusters no espaço de características revela a estrutura dos agrupamentos e fornece informações visuais sobre como as amostras são divididas.

## Conclusão
Este trabalho acadêmico demonstra a aplicação bem-sucedida do modelo K-Means para agrupar dados do conjunto de dados Iris com base em suas características. O K-Means é uma técnica valiosa para a análise de clusters em dados e pode ser aplicado a uma variedade de domínios.

A análise do índice de silhueta e das métricas de distância contribui para uma compreensão mais profunda da qualidade do agrupamento, enquanto a visualização dos clusters auxilia na interpretação dos resultados.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score

# Carregar a base de dados iris
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# Selecionar as features a serem utilizadas
features = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]]

# Definir o número de clusters
n_clusters = 3

# Instanciar o modelo K-Means
model = KMeans(n_clusters=n_clusters)

# Treinar o modelo
model.fit(features)

# Obter os rótulos dos clusters
labels = model.labels_

# Número de amostras em cada cluster
cluster_counts = np.bincount(labels)

# Criar um DataFrame com as coordenadas dos centróides
centroids_df = pd.DataFrame(model.cluster_centers_, columns=features.columns)

# Exibir estatísticas dos clusters
for i in range(n_clusters):
    print(f"Cluster {i} - Número de Amostras: {cluster_counts[i]}")
    print(f"Centróide {i}:")
    print(centroids_df.iloc[i])
    print("\n")

# Visualizar os clusters
plt.scatter(features["sepal_length"], features["sepal_width"], c=labels, cmap="Set1")
plt.show()
```
